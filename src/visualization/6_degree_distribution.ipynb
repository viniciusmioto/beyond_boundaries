```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import Counter

FILES_PATH = "../../data/graphs/full"
```

```python
def plot_degree_distribution(G, log_log=True):
    """
    Plots the degree distribution of a NetworkX graph using a Seaborn scatterplot.

    Parameters:
        G (networkx.Graph): The input graph.
        log_log (bool): If True, uses log-log scales for both axes (default True).

    Returns:
        plt.Figure: The matplotlib figure containing the degree distribution plot.
    """
    # Apply a research paper style theme
    sns.set_theme(context="paper", style="whitegrid")
    
    # Compute the degree sequence and frequency distribution
    degrees = [d for n, d in G.degree()]
    degree_count = Counter(degrees)
    
    # Create a DataFrame for plotting
    data = pd.DataFrame({
        "Degree": list(degree_count.keys()),
        "Frequency": list(degree_count.values())
    })
    
    # Create the figure and axes
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot the degree distribution using Seaborn's scatterplot
    sns.scatterplot(data=data, x="Degree", y="Frequency", ax=ax)
    
    # Set axes to log scale if required
    if log_log:
        ax.set_xscale("log")
        ax.set_yscale("log")
 
    ax.set_xlabel("Degree")
    ax.set_ylabel("Frequency")
    
    ax.set_title("Degree Distribution (Number of Collaborations)")
    plt.tight_layout()
    
    return fig
```

```python
collabnet_br = nx.read_gexf(f"{FILES_PATH}/collabnet_br.gexf")
```

```python
plot_degree_distribution(collabnet_br)
plt.show()
```
