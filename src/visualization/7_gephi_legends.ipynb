```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

COUNTRY_PALLETE = {
    "BR": "#0aad2d",
    "US": "#184496",
    "GB": "#c658de",
    "CA": "#ED2577",
    "NL": "#ff6f00",
    "CN": "#df000a",
    "IT": "#82D1B5",
    "DE": "#6c09ba",
    "IN": "#dea323",
    "ES": "#BDA80D",
    "FR": "#5fb3cf",
    "AU": "#6876FF",
    "Others": "#c0c0c0"
}

def plot_country_legend(country_palette=COUNTRY_PALLETE):
    # Ensure every color starts with '#'
    palette = {
        code: (color if color.startswith('#') else '#' + color)
        for code, color in country_palette.items()
    }
    
    n_items = len(palette)
    # Adjust figure height based on the number of items
    fig_height = n_items * 0.5
    fig, ax = plt.subplots(figsize=(3, fig_height))
    ax.axis('off')  # Hide the axes
    
    square_size = 0.3  # Relative size for the squares
    x_offset = 0.01     # X starting coordinate
    text_offset = x_offset + square_size + 0.1  # X coordinate for the text

    for i, (code, color) in enumerate(palette.items()):
        # Calculate y coordinate (top row at highest y)
        y = n_items - i - 0.5
        # Draw the colored square
        rect = Rectangle((x_offset, y - square_size/2), square_size, square_size,
                         facecolor=color, edgecolor='black')
        ax.add_patch(rect)
        # Place the country code text next to the square
        ax.text(text_offset, y, code, va='center', fontsize=10)
    
    # Set limits to tightly fit the legend items
    ax.set_xlim(0, 3)
    ax.set_ylim(0, n_items)
    plt.savefig("../../results/aux/international_legend.svg", format="svg")
    plt.show()

# Example usage:
plot_country_legend()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

COLOUR_PALETTE = {
    "AI": "#D32F2F",  # Artificial Intelligence
    "Vision & Recognition": "#1976D2",  # Computer Vision and Pattern Recognition
    "Info Systems": "#2E7D32",  # Information Systems
    "Theory & Math": "#6A1B9A",  # Computational Theory and Mathematics
    "Networks & Communications": "#FF6F00",  # Computer Networks and Communications
    "CS Applications": "#00838F",  # Computer Science Applications
    "Software": "#0ad600",  # Software
    "Signal Processing": "#2CCFDB",  # Signal Processing
    "HCI": "#CE59D4",  # Human-Computer Interaction
    "Hardware & Arch": "#455A64",  # Hardware and Architecture
    "Graphics & Aided-Design": "#D6914B",  # Computer Graphics and Computer-Aided Design
}

def plot_subfield_legend(colour_palette=COLOUR_PALETTE):
    # Ensure every color string starts with '#'
    palette = {
        key: (value if value.startswith('#') else '#' + value)
        for key, value in colour_palette.items()
    }
    
    # Sort the palette alphabetically by the subfield names (keys)
    sorted_items = sorted(palette.items(), key=lambda item: item[0])

    n_items = len(sorted_items)
    # Adjust figure height based on the number of items
    fig_height = n_items * 0.5
    fig, ax = plt.subplots(figsize=(4, fig_height))
    ax.axis('off')  # Hide the axes
    
    square_size = 0.3  # Size of the square
    x_offset = 0.1     # X starting coordinate for the square
    text_offset = x_offset + square_size + 0.1  # X coordinate for the text

    for i, (subfield, color) in enumerate(sorted_items):
        # Calculate y coordinate so that the first item is at the top
        y = n_items - i - 0.5
        # Draw the colored square
        rect = Rectangle((x_offset, y - square_size/2), square_size, square_size,
                         facecolor=color, edgecolor='black')
        ax.add_patch(rect)
        # Place the subfield text next to the square
        ax.text(text_offset, y, subfield, va='center', fontsize=10)
    
    # Set limits to ensure the legend is nicely fit
    ax.set_xlim(0, 4)
    ax.set_ylim(0, n_items)
    plt.savefig("../../results/aux/subfields_legend.svg", format="svg")
    plt.show()

# Example usage:
plot_subfield_legend()
```
